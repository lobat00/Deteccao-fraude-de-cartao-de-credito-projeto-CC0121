## 1. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", None)
```

## 2. Carregar os dados

```python
df = pd.read_csv("../data/creditcard.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum().sum()  # checar valores ausentes
```

## 3. Distribuição da variável alvo (`Class`)

```python
class_counts = df["Class"].value_counts()
class_pct = df["Class"].value_counts(normalize=True) * 100
print(class_counts)
print(class_pct)
```

```python
sns.countplot(data=df, x="Class")
plt.title("Distribuição de Classes (0 = Legítima, 1 = Fraude)")
plt.yscale("log")
plt.show()
```

## 4. Análise de `Amount`

```python
df.groupby("Class")["Amount"].describe()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.boxplot(data=df, x="Class", y="Amount", ax=axes[0])
axes[0].set_title("Amount por Classe")

sns.histplot(data=df, x="Amount", hue="Class", bins=50, log_scale=(False, True), ax=axes[1])
axes[1].set_title("Distribuição de Amount")
plt.tight_layout()
plt.show()
```

## 5. Análise de `Time`

```python
df["Hour"] = (df["Time"] // 3600) % 24

fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(data=df, x="Hour", hue="Class", bins=24, multiple="dodge", stat="density", common_norm=False)
plt.title("Distribuição de Transações por Hora (normalizada por classe)")
plt.show()
```

## 6. Correlações entre features

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(16, 12))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Matriz de Correlação")
plt.show()
```

```python
# Correlação de cada feature com Class, ordenada
corr["Class"].sort_values(ascending=False)
```
